```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Q. 100  .  ,  ,    

# 0.   ()
np.random.seed(42)
days = 100
data = {
    'Date': pd.date_range(start='2024-01-01', periods=days),
    'Category': np.random.choice(['Electronics', 'Clothing', 'Home', 'Books'], days),
    'Sales': np.random.randint(1000, 5000, days),
    'Profit': np.random.randint(100, 1000, days)
}
df = pd.DataFrame(data)
df = df.set_index('Date')




# --- Mission 1:   (Line Plot) ---
# figsize   
plt.figure() #   
df['Sales'].plot(kind='line', figsize=(10, 5), title='Daily Sales Trend')
plt.show()

# --- Mission 2:    (Bar Plot) ---
plt.figure()
# groupby   ->   
df.groupby('Category')['Sales'].sum().plot(kind='bar', color='skyblue', title='Sales by Category')
plt.show()

# --- Mission 3:    (Scatter Plot) ---
plt.figure()
df.plot(kind='scatter', x='Sales', y='Profit', c = 'red', alpha = 0.5, s = 50, title='Sales vs Profit')
    #s(size) df['profit']          
plt.show()
```
