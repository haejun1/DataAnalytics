```python
import pandas as pd

data = {
    'price': [2500, 15000, 5000],
    'stock': [10, 2, 5]
}
df = pd.DataFrame(data, index = ['apple', 'melon', 'cherry'])
```

```python
#iloc으로 마지막 행, 마지막 열 값 추출해보기
print(df.iloc[-1,-1])

#boolean indexing으로 재고가 5이상, 가격이 3000이상인 과일 index 추출
fbi = (df['stock'] >=5) & (df['price'] >3000) #괄호 주의
print(df[fbi].index)

#인덱스에 'e'문자가 포함된 과일들의 가격 평균 구하기
fs = df.index.str.contains('e')
fm = df.loc[fs, 'price'].mean()
print(fm)

#loc를 이용한 데이터 수정 (chained Indexing 주의)
print("인상 전 멜론 : ",df.loc['melon'],'\n')
df.loc['melon','price'] *= 1.2
print("인상 후 멜론 : ", df.loc['melon'])
```
