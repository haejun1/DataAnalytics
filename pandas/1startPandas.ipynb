```python
import pandas as pd

#1. 과일 데이터 생성 (Dictionary to DataFrame)
#인덱스를 설정한다
data = {
    'price' : [100, 150, 200, 250, 300],
    'stock': [20, 30, 15, 10, 5]
}
df = pd.DataFrame(data, index=['apple', 'banana', 'orange', 'grape', 'peach'])

print(df.head(),"\n")

#데이터 요약 (데이터 타입, 결측치 확인)
print(df.info(),"\n")

#기본 통계량 확인(평균, 표준편차, 최소값, 최대값 등)
print(df.describe(),"\n")

#데이터 연산 (Vectorized Operation)
mean_price = df['price'].mean()
print(f"평균 가격: {mean_price:.1f}원\n")

# CSV로 저장해보기 (I/O)
df.to_csv('fruit_data.csv')
print("\n'fruit_data.csv' 파일이 저장되었습니다.")
```

```python
# 특정 데이터 접근 : loc 사용
print(df.loc['apple'],'\n')

#df.loc[행, 열] : 특정 행의 특정 열에 접근
print(df.loc['apple','price'],'\n')

df['stock'] = df['stock']+5

print(df.describe())
```

```python
#인덱스의 재구성 (set_index, reset_index)

#원래 과일이름들이 index로 설정되어 있었음 -> 리셋으로 초기화
print(df.head(),"\n")
df =df.reset_index()
print("인덱스 초기화 \n",df.head(),"\n")

#다시 과일이름을 index(PK)로 설정
df = df.set_index('index')
print("인덱스 재설정 \n",df.head(),"\n")
```

```python
# Axis 축의 개념

#axis=0 : 가로 방향 (인덱스 기준으로 연산--)
#axis=1 : 세로 방향 (컬럼 기준으로 연산||)

print(df.sum(axis=0),"\n")   #각 (가로)열의 합계
print(df.sum(axis=1),"\n")   #각 (세로)행의 합계
```
