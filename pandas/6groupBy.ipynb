```python
import pandas as pd
import numpy as np

# 영업팀 실적 데이터
data = {
    'Team': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
    'Employee': ['Kevin', 'Tom', 'Jerry', 'Mike', 'Sarah', 'Bob', 'Alice', 'Jenny', 'David', 'Chris'],
    'Sales': [1200, 1500, 800, 2000, 2100, 1800, 500, 400, 300, np.nan],
    'Bonus_Target': [1000, 1000, 1000, 2000, 2000, 2000, 500, 500, 500, 500]
}

df = pd.DataFrame(data)

print("--- 원본 영업 데이터 ---")
print(df)
```

```python
#1. 팀별 성과표 만들기
df_team_score = df.groupby('Team', as_index=False).agg(
    {'Sales' : ('sum','mean'),
     'Employee' : 'count'}
)
print(df_team_score)
```

```python
#2. 결측치를 평균으로 변환
df['Sales'] = df['Sales'].fillna(df.groupby('Team')['Sales'].transform('mean'))
    #결측치를 채우겠다 -> 뭐로?
    #'Team'으로 그룹화 한것을 'Sales'의 평균으로
print(df)
```

```python
#3. 팀 내 기여도 계산
df['Share'] = df['Sales'] / df.groupby('Team')['Sales'].transform('sum')
print(df)
```

```python
#4. 팀 평균 매출 1000 미만인 팀 데이터 삭제
df_survivor = df.filter(lambda df : df['Sales'].mean() > 1000)
print(df_survivor)
```
