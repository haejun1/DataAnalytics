```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

df = pd.read_csv("kaggle_dataset\mobilePrice\mobilePrice.csv", sep=",")

df.head()
```

```python
#1. 미사용 컬럼 제거
df = df.drop(columns=['id', 'm_dep'])
```

```python
#2. 결측치 처리
df.isna().sum()
```

```python
#3. 이상치 처리
```

```python
#4. Feature Engineering
battery = df['talk_time'] / df['battery_power']
df['battery'] = battery * 100

camera = (df['fc'] + df['pc']) / 2
df['camera'] = camera

pixel = df['px_height'] * df['px_width']
pixel = pixel.replace(0, np.nan)
display_size = (df['sc_h'] + df['sc_w']) * 100000
display = display_size / pixel
df['display'] = display
```

```python
#5. 정규화

num_cols = [
    'battery',
    'clock_speed',
    'camera',
    'display',
    'int_memory',
    'mobile_wt',
    'n_cores',
    'ram'
]

for col in num_cols:
    mean = df[col].mean()
    std = df[col].std()
    df[col] = (df[col] - mean) / std

df.head()
```
