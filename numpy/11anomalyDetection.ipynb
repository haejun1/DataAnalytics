```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.decomposition import PCA

df = pd.read_csv('marketing_worekd.csv')
features = ['Income', 'Age', 'Total_Spent', 
            'Total_Campaigns', 'NumWebVisitsMonth', 'Mnt_Premium', 'Education_Level']
x = df[features]
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
#Isolation Forest 모델 생성 및 학습
model = IsolationForest(contamination=0.03, random_state=42)
    #contamination=0.03, 전체 데이터의 3%를 이상치로 가정
df['Anomaly'] = model.fit_predict(x_scaled)
df['Anomaly_Score'] = model.decision_function(x_scaled)
    #이상치 점수 추출
```

```python
#PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
pca_df = pd.DataFrame(x_pca, colums=['PC1','PC2'])
pca_df['Anomaly'] = df['Anomaly']
    #pca_df라는 새로운 df에 Anomaly라는 새로운 컬럼을 넣고
    #Isolation Forest를 거친 df['Anomaly']값을 새로운 칼럼에 넣는다
    #즉 pca_df에 차원축소한 점들의 위치가 있고 그것을 Anomaly칼럼이 이상치를 판단해준다

plt.figure(figsize=(10,6))
sns.scatterplot(x='PC1',y='PC2',hue='Anomaly',data=pca_df,palette={1:'blue',-1:'red'},alpha=0.6)
    #hue=색깔기준, 이상치 = -1
plt.title('Isolation Forest Anomaly Detection (red=Anomaly)')
plt.show()
print(df.groupby('Anomaly')[features].mean())
```
